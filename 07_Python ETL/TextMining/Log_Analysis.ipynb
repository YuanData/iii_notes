# 計算詞頻

```python

from  sklearn.feature_extraction.text  import  CountVectorizer  
  
#將文本中的詞語轉換為詞頻矩陣  
vectorizer = CountVectorizer()  
print(len(allSentence))
#計算個詞語出現的次數  
X = vectorizer.fit_transform(allSentence)  
#獲取詞袋中所有文本關鍵詞  
word = vectorizer.get_feature_names()  
## print(word)  
#查看詞頻結果  
## print(X.toarray())
```

# 計算Tf-Idf

```python
from  sklearn.feature_extraction.text  import  TfidfTransformer  
  
#  
transformer = TfidfTransformer()  

 
#將詞頻矩陣X統計成TF-IDF值  
tfidf = transformer.fit_transform(X)  
#查看數據結構 tfidf[i][j]表示i類文本中的tf-idf權重  
print(tfidf.toarray())
```

# 計算文件相似度

```python
from sklearn.metrics.pairwise import cosine_similarity
## print(tfidf)



sepList = []
for i in range(len(allSentence)):
    array = cosine_similarity(tfidf[i:i+1], tfidf)[0]
    sepSentence = []
    for j in range(len(allSentence)):
        ### 決定相似度較高達多少一組，此處為經驗法則，看組合是否能滿足需求。
        if array[j] > 0.5:
            ##print(allSentence[j])
            sepSentence.append(allSentence[j])
        sepList.append(sepSentence)

```

```python
### 可觀察 每篇向量 X 每篇向量 對角線會為一，主要原因是自己 X 自己相似度最高
arrayTest = cosine_similarity(tfidf, tfidf)
print(arrayTest)
```

```python
finalList = []
### 此處用來過濾 相似度陣列中完全一樣的組合。
for k in sepList:
    if k not in finalList:
        finalList.append(k)      
```

# 分析相似度0.5以上群聚較高 跟 群聚較低的問句組合

```python
### 相似度的分水嶺取決於問句本身的架構是否完整。
### 取組合中，高相似度，且群集量較中階的組合，稍有問句的雛形。
### 由下方相似度較高的組合中針對每組找出關鍵的目的 跟 特定字眼 揣摩使用者想要問的內容。
### 一開始不用word count 來解決，主要是先抓大方向先收斂一次問句。
countHigh = 0
impHigh = []
for imp in finalList:  
    if len(imp) > 2 and len(imp) < 10:
        countHigh+=1
        impHigh.append(imp)
        print(imp)
print(countHigh)
```

```python
### 取組合中，高相似度，且群集量較低的組合，量也是最大的，主要是單純名詞或著隨興問答就多。
### 單詞類的問答，比較虛無飄渺，故以字典方式回應。
countLow = 0
impLow = []
for imp in finalList:  
    if len(imp) > 0 and len(imp) < 2:
        countLow+=1
        impLow.append(imp)
        ##print(imp)
##print(countLow)


### Setup Stop Word
stopwordset = set()
with open('con_stopwords.txt','r',encoding='utf-8') as sw:
    for line in sw:
        stopwordset.add(line.strip('\n'))

stopWordList = []
for stopWord in stopwordset:     
    stopWordList.append(stopWord)
    
### Word Count and filter by stop word
wordDict = {}
for lowList in impLow:
    for sentence in lowList:
        wordList = sentence.split(' ')[:-1]
        ##print(wordList)
        for word in wordList:
            ### filter by stop word
            if word not in stopWordList:
                if word not in wordDict.keys():
                    wordDict[word]=1
                else:
                    ##print(word)
                    wordDict[word]+=1
                    ##print(wordDict[k])
                    ##print('------')
countKeyWord = 0
for i in wordDict:
    countKeyWord+=1
    ##print(i)
print(countKeyWord)

### Sorting the wordDict and build the hot word in dict 
dict= sorted(wordDict.items(), key=lambda d:d[1], reverse = True)

for j in dict:
    print(j)
```

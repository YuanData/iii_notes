```python
##  mysql 調資料顯示在機器人上
##  球隊/球員整個的打擊/投球成績，包括：

##  1.投手的出賽場次所屬球隊、W、L、ERA、G、GS、QS、IP、R、ER、HR、BB、SO、WHIP、GO/AO、GIDP、S%、RS/9、K/9、BB/9、HR/9
##  2、打者數據：[team]、G、ab、h、2b、3b、hr、rbi、bb、so、sb、avg、obp、slg、ops、go/ao
##  3、球隊團隊：W、L、AVG、OBP、SLG、OPS、ERA、WHIP
##  對象：
##  1.投手下次對手球隊的投球成績
##  2.各個打者對下次對手的打擊成績
##  3.球員近三場的成績

##  一、投手SQL語法
##  select [*] from pitchergamelog where game_pk in (select game_pk from gamedetail where game_date 
##  between [20160101] and [now()]) and player_id = [547888];

## 秀1:round(sum(er)*9/(sum(truncate(ip,0)) + sum(ip-truncate(ip,0))/0.3),2) ERA 防禦率
## 秀2:round((sum(h)+sum(bb))/(sum(truncate(ip,0)) + sum(ip-truncate(ip,0))/0.3),2) WHIP 每局被上壘率
## 秀3:sum(s)/sum(np) StrikeRate 好球率
## 秀4:sum(so)/sum(bb) KperBB 三振四壞比率
## 秀5:sum(go)/sum(ao) GOperAO 滾飛比


##  註：[*] = sum(w) Win, sum(l) Lose, round(sum(er)*9/(sum(truncate(ip,0)) + sum(ip-truncate(ip,0))/0.3),2) ERA , 
##           sum(g) Game, sum(gs) GameStart, sum(truncate(ip,0)) + sum(ip-truncate(ip,0))/0.3 IP, sum(r) Run, 
##           sum(er) EarnedRun, 
##           sum(hr) HomeRun, 
##           sum(bb) Walk, 
##           sum(so) Strikeout, 
##           round((sum(h)+sum(bb))/(sum(truncate(ip,0)) + sum(ip-truncate(ip,0))/0.3),2) WHIP, 
##           sum(go)/sum(ao) GOperAO, sum(gidp) DoublePlay, sum(s)/sum(np) StrikeRate, 
##           sum(so)*9/(sum(truncate(ip,0)) + sum(ip-truncate(ip,0))/0.3) K9, 
##           sum(bb)*9/(sum(truncate(ip,0)) + sum(ip-truncate(ip,0))/0.3) BB9, 
##           sum(hr)*9/(sum(truncate(ip,0)) + sum(ip-truncate(ip,0))/0.3) HR9

##  求QS語法： select count(*) QS from pitchergamelog where game_pk in (select game_pk from gamedetail where game_date 
##            between 20160101 and now()) and gs = 1 and ip >= 6 and r <= 3 and player_id = [547888];

##  PS. IP 欄位要再以其他語法把分數改為小數，ex.10.66667 需改成 10.2 、 9.33333 需改成 9.1

##  二、打者SQL語法
##  select [*] from hittinggamelog where game_pk in (select game_pk from gamedetail where game_date between [20160101] and [now()]) and player_id = [444876];

##  註：[*] = count(*) Game, sum(ab) AB, sum(h) Hit, sum(d) 2B, sum(t) 3B, sum(hr) HomeRrun, sum(rbi) RBI, sum(bb) Walk, 
##            sum(so) Strikeout,
##            sum(sb) StolenBase, 
##            sum(h)/sum(ab) AVG, 
##           (sum(h)+sum(bb)+sum(hbp))/(sum(ab)+sum(bb)+sum(hbp)+sum(sf)) OBP, 
##           sum(tb)/sum(ab) SLG, 
##           (sum(h)+sum(bb))/(sum(ab)+sum(bb)+sum(sf))+sum(tb)/sum(ab) OPS, 
##           sum(go)/sum(ao) GOperAO

## 秀1:sum(h)/sum(ab) AVG 打擊率
## 秀2:(sum(h)+sum(bb)+sum(hbp))/(sum(ab)+sum(bb)+sum(hbp)+sum(sf)) OBP 上壘率
## 秀3:sum(tb)/sum(ab) SLG 長打率
```
